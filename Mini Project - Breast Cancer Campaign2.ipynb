```python
import pandas as pd
import numpy as np
df = pd.read_csv('insurance.csv')
df
```

```python
df = df.dropna()
df
```

```python
from sklearn.preprocessing import LabelEncoder

df["sex"] = LabelEncoder().fit_transform(df[["sex"]])
```

```python
df["smoker"] = LabelEncoder().fit_transform(df[["smoker"]])
```

```python
df["region"] = LabelEncoder().fit_transform(df[["region"]])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df_scaled = scaler.fit_transform(df.to_numpy())

df_scaled = pd.DataFrame(df_scaled, columns= ['age', 'sex', 'bmi', 'children', 'smoker', 'region', 'charges'])

df_scaled
```

```python
X = df_scaled.iloc[:, 0:6]
y= df.iloc[:, 6:7]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_validate, X_test, y_validate, y_test = train_test_split(X, y, test_size=0.5)
```

```python
from sklearn.ensemble import RandomForestRegressor

modelLR = RandomForestRegressor(criterion = "mse").fit(X_train, y_train)
score = modelLR.score(X_test, y_test)
y_pred = modelLR.predict(X_test)
print(score)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0).fit(X_train, y_train)
score = regressor.score(X_test, y_test)
y_pred = regressor.predict(X_test)
print(score)
```
