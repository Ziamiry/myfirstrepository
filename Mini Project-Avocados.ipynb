```python
import pandas as pd
import numpy as np
df = pd.read_csv("avocado.csv")
df
```

```python
df.dropna()
```

```python
df.drop(['region', 'Date', 'Unnamed: 0'], axis=1, inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder

df["type"] = LabelEncoder().fit_transform(df[["type"]])
df
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df_scaled = scaler.fit_transform(df.to_numpy())

df_scaled = pd.DataFrame(df_scaled, columns= ['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags',
       'Small Bags', 'Large Bags', 'XLarge Bags', 'type', 'year'])

df_scaled
```

```python
X = df_scaled.iloc[:, 1:12]
y= df.iloc[:, 0:1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_validate, X_test, y_validate, y_test = train_test_split(X, y, test_size=0.5)
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors = 8).fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
scores = []
results = 0
best_score = 0
neighbors = range(1,15)

for i in neighbors:
    
    knn = KNeighborsRegressor(n_neighbors = i).fit(X_train,y_train)

    results = knn.score(X_test,y_test)
    scores.append(round(results,2))
    if  results > best_score:
        best_score = results 
        best_k = i
        bestmodel = knn

    print(scores)
    print (best_k)
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbors, scores)
```

```python
accuracy = bestmodel.score(X_validate, y_validate)

print("The best model has an accuray of:", round(accuracy,2) )
```

```python
from sklearn.linear_model import LinearRegression

modelLR = LinearRegression().fit(X_train, y_train)
score = modelLR.score(X_test, y_test)
print(score)
```

```python
from sklearn.ensemble import RandomForestRegressor

modelLR = RandomForestRegressor(criterion = "mse").fit(X_train, y_train)
score = modelLR.score(X_test, y_test)
y_pred = modelLR.predict(X_test)
print(score)
```

```python
from sklearn.metrics import mean_squared_error
print ("Accuracy:", score)
print ("MSE:" + str(mean_squared_error(y_test, y_pred)))
```
